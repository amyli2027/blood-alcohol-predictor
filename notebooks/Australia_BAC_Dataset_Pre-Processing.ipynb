```python
import pandas as pd
```

```python
bac_dataset = pd.read_csv("../data/raw/bac_dataset.csv")
bac_dataset_sorted = bac_dataset.sort_index()
bac_dataset_sorted
```

```python
australia_bac_dataset = bac_dataset_sorted.loc[:, 'ID_AUST': ]
australia_bac_dataset
```

```python
australia_bac_dataset = australia_bac_dataset.dropna()
australia_bac_dataset = australia_bac_dataset.drop('ID_AUST', axis = 1)
australia_bac_dataset
```

```python
australia_bac_dataset.info()
```

```python
australia_bac_dataset['Wght_AUS'] = australia_bac_dataset['Wght_AUS'].astype(float)
australia_bac_dataset['Height'] = australia_bac_dataset['Height'].astype(int)
australia_bac_dataset['Age'] = australia_bac_dataset['Age'].astype(int)
australia_bac_dataset['Wine'] = australia_bac_dataset['Wine'].astype(int)
australia_bac_dataset['1hr-BAC'] = australia_bac_dataset['1hr-BAC'].astype(float)
australia_bac_dataset.info()
```

```python
gender_one_hot = pd.get_dummies(australia_bac_dataset['Gend_AUS'], prefix='Gender', drop_first=True)
australia_bac_dataset = pd.concat([australia_bac_dataset, gender_one_hot.astype(int)], axis=1)
australia_bac_dataset = australia_bac_dataset.drop('Gend_AUS', axis=1)
australia_bac_dataset
```

```python
australia_bac_dataset['Wght_AUS'] = australia_bac_dataset['Wght_AUS'] * 453.6
australia_bac_dataset = australia_bac_dataset.rename(columns={'Wght_AUS': 'Bodyweight_grams'})
australia_bac_dataset['Bodyweight_grams']
```

```python
australia_bac_dataset['Wine'] = australia_bac_dataset['Wine'] * 9.47
australia_bac_dataset = australia_bac_dataset.rename(columns={'Wine': 'Ethanol_grams'})
australia_bac_dataset['Ethanol_grams']
```

```python
australia_bac_dataset.to_csv('australia_bac_dataset_processed.csv', index=False)
```
